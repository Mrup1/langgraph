```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
class MyState(TypedDict):
    name: str
    rollno: int
    
```

```python
def info(state: MyState) -> str:
    return f"{state['name']} has roll number {state['rollno']}"
```

```python
workflow = StateGraph(MyState)
workflow.add_node('get_info',info)
workflow.add_edge(START, "get_info")
workflow.add_edge("get_info", END)
```

```python
graph = workflow.compile()
```

```python
# create a new workflow that wraps the existing `info` node to return a dict (required by StateGraph)
workflow2 = StateGraph(MyState)

def info_wrapper(state: MyState) -> dict:
    # call the original function (it returns a string) but return the state dict as the node update
    _ = info(state)
    return {"name": state["name"], "rollno": state["rollno"]}

workflow2.add_node("get_info_fixed", info_wrapper)
workflow2.add_edge(START, "get_info_fixed")
workflow2.add_edge("get_info_fixed", END)

graph2 = workflow2.compile()
output = graph2.invoke({"name": "Alice", "rollno": 42})
print(output)
print(output)
```
