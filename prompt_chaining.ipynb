```python
from langgraph. graph import StateGraph , START , END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash", 
    temperature=0.7
)
```

```python
class MyState(TypedDict):
    topic: str
    easy:str
    evaluation: str
    number: int
    
```

```python
def easy_node(state: MyState) -> MyState:
    topic = state['topic']
    prompt = f'write a good easy on the topic - {topic}'
    easy = llm.invoke(prompt).content
    state['easy'] = easy
    return state
```

```python
def evaluation_node(state: MyState) -> MyState:
    easy = state['easy']
    prompt = f'evaluate the following easy - {easy}'
    evaluation = llm.invoke(prompt).content
    state['evaluation'] = evaluation
    return state
```

```python
def number_node(state: MyState) -> MyState:
        easy = state['easy']
        prompt = f'give this easy number out of 10 - {easy}'
        number = llm.invoke(prompt).content
        state['number'] = int(number)
        return state
```

```python
graph = StateGraph(MyState)
graph.add_node('easy',easy_node)
graph.add_node('evaluation',evaluation_node)
graph.add_node('number',number_node)
graph.add_edge(START,'easy')
graph.add_edge('easy','evaluation')
graph.add_edge('evaluation','number')
graph.add_edge('number',END)    


```

```python
workflow = graph.compile()
```

```python
output = workflow.invoke({'topic':'agentic ai through lang graph'})
```
