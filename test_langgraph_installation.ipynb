```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
# 1. Define the structure of the data passing through your graph
class State(TypedDict):
    message: str

```

```python
# 2. Create a simple node function that modifies the data
def hello_node(state: State):
    print("--- Inside Hello Node ---")
    return {"message": state["message"] + " World!"}

```

```python
# 3. Build the graph structure
workflow = StateGraph(State)
workflow.add_node("greeting_node", hello_node)

# 4. Connect the dots with edges
workflow.add_edge(START, "greeting_node")
workflow.add_edge("greeting_node", END)

# 5. Compile it into an executable app
app = workflow.compile()

# Run it!
output = app.invoke({"message": "Hello"})
print("Final Output:", output)
```
